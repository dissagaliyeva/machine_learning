```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set()
```

```python
# load the datasets
all_states = pd.read_csv('datasets/all_states.csv')
swing_states = pd.read_csv('datasets/swing_states.csv')

print(all_states.head(3))
print(swing_states.head(3))
```

```python
rep_dem = swing_states[['state', 'dem_votes', 'rep_votes']]

states = np.unique(swing_states.state)
# plot swing states with bar chart
plt.title('Red & Blue states')
plt.bar(states, height=[swing_states[swing_states.state == x]['rep_votes'].sum() / 1e+6 for x in states],
        label='republican', color='red', alpha=0.5, )
plt.bar(states, height=[swing_states[swing_states.state == x]['dem_votes'].sum() / 1e+6 for x in states],
        label='democrat', color='blue', alpha=0.5)
plt.legend()
plt.show()
```

### Empirical Cumulative Distribution Function (ECDF)
It is a way to represent the spread of the data when normal probability distribution or bar plots cannot help.
ECDF shows the probability on the y-axis, thus we can see a clearer picture.
